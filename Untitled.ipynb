```python
import pandas as pd
import numpy as np
import pickle

import lda
import pandas as pd
import numpy as np
import lda.datasets
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.cluster.hierarchy import ward, dendrogram

#from sklearn.lda import 
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.decomposition import TruncatedSVD 
from sklearn.preprocessing import normalize 
```

```python
df = pd.read_pickle('pkl/df_stop_noun.pkl')
df.head(3)
```

```python
n_samples = 2000
n_features = 10000
n_topics = 10
n_top_words = 20

tfidf_vec = TfidfVectorizer(ngram_range=(1, 3), max_df=0.9, min_df=2, 
                max_features=n_features, norm='l2', use_idf=True, smooth_idf=True, sublinear_tf=False)

tfidf_vec_prep = tfidf_vec.fit_transform(df['resume_nouns'])
```

```python
tfidf_vec_prep.shape
```

```python
svd = TruncatedSVD(n_components=2000)
embedded_coords = normalize(svd.fit_transform(tfidf_vec_prep), norm='l1')
print(embedded_coords.shape)
```

```python
import os
import nltk

class MySentences2(object):
     def __init__(self, dirname):
            self.dirname = dirname
 
     def __iter__(self):
         for fname in os.listdir(self.dirname):
                for line in open(os.path.join(self.dirname, fname), encoding='latin-1'):
                    word=line.lower().split()
                    yield word
                    
sentences = MySentences2('/Users/bryant/nltk_data/corpora/gutenberg') 
model = gensim.models.Word2Vec(sentences, size=100, window=5, min_count=1, workers=4)
```

```python
dist = 1 - cosine_similarity(tfidf_vec_prep)

linkage_matrix = ward(dist) #define the linkage_matrix using ward clustering pre-computed distances
```

```python
fig, ax = plt.subplots(figsize=(15, 20)) # set size
ax = dendrogram(linkage_matrix, orientation="right");

plt.tick_params(\
    axis= 'x',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom='off',      # ticks along the bottom edge are off
    top='off',         # ticks along the top edge are off
    labelbottom='off')

plt.tight_layout() #show plot with tight layout

#uncomment below to save figure
plt.savefig('ward_clusters.png', dpi=200) #save figure as ward_clusters
```
